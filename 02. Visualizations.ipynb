## A. Univariate visualizations

Nous utiliserons les librairies `matplotlib` et `seaborn` pour générer les graphiques.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### –– **Variable qualitative**  

- **Pie chart**

```python
#Plot a pie chart
data["main_category"].value_counts().plot(kind='pie').axis('equal');
```

- **Bar chart**

```python
#Set normalize to True to visualize percentages on y-axis
data["main_category"].value_counts(normalize=True).plot(kind='bar');
```

### –– **Variable quantitative**  

Pour l'histogramme, il existe des règles pour déterminer le nombre optimal de classes. Par exemple, la **<mark>règle de Sturges</mark>** considère comme nombre optimal de classes :

$$\\k = [1 + log_2 (n)]$$

où n est la taille de l'échantillon.

```python
#Plot a histogram with k bins
data["usd_goal_real"].hist(bins=13);
```

Zoomons là où semble se concentrer les valeurs pour mieux en visualiser la distribution.

```python
#Zoom under 250000 + add stats
med = data['usd_goal_real'].median()
mean = data['usd_goal_real'].mean()

fig, ax = plt.subplots()
data[data.usd_goal_real < 250000]["usd_goal_real"].hist(bins=13)
ax.vlines(x=med, ymin=0, ymax=100000, color='red')
ax.vlines(x=mean, ymin=0, ymax=100000, color='green')
plt.show()
```

### –– Visualize stats for each category of a categorical variable

```python
import matplotlib.pyplot as plt

mean = data['usd_goal_real'].mean()
#MMM for each category of a categorical variable
for main_category in data["main_category"].unique():
    subset = data[data.main_category == main_category] # Création du sous-échantillon
    print("-"*20)
    print(main_category)
    print("moy:\n",subset['usd_pledged_real'].mean())
    print("med:\n",subset['usd_pledged_real'].median())
    print("mod:\n",subset['usd_pledged_real'].mode())
    print("var:\n",subset['usd_pledged_real'].var(ddof=0))
    print("std:\n",subset['usd_pledged_real'].std(ddof=0))
    fig, ax = plt.subplots()
    subset["usd_pledged_real"].hist() # Crée l'histogramme
    ax.vlines(x=mean, ymin=0, ymax=40000, color='green')
    plt.show() # Affiche l'histogramme
```

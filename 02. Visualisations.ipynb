### Coefficient de corrélation de Pearson $r$

La corrélation de Pearson est linéaire. Elle est compris entre -1 et 1.

- $r$ ∈ ]0;1] : x et y sont corrélés positivement (si x augmente, y augmente)
- $r$ = 0 : x et y ne sont pas corrélés
- $r$ ∈ [-1;0[ : x et y sont corrélés négativement (si x augmente, y diminue et réciproquement)

```python
Image(filename='img/pearson.jpg', width="600")
```

Dans la 3ème figure, on constate que $r$ vaut 0.17. Pourtant, x et y semblent corrélés. Le fait est que leur relation n'est pas linéaire.

```python
plt.plot(data["usd_pledged_real"], data["backers"], 'o', alpha=0.5)
plt.xlabel("usd_pledged_real")
plt.ylabel("backers")
plt.show()
```

ATTENTION : Quand on découvre une corrélation proche de |0.5|, on peut se demander si les variables sont à déclarer corrélées ou non corrélées. Dans ce cas-là, on passera par des tests statistiques en formulant une hypothèse de corrélation à vérifier selon un degré de signicativité (p-value) défini.

```python
bins = 100000 #binning

groupes = [] # va recevoir les données agrégées à afficher

# on calcule des tranches allant de 0 au solde maximum par paliers de taille taille_classe
tranches = np.arange(0, max(data["usd_pledged_real"]), bins)
tranches += bins/2 # on décale les tranches d'une demi taille de classe
indices = np.digitize(data["usd_pledged_real"], tranches) # associe chaque solde à son numéro de classe

for ind, tr in enumerate(tranches): # pour chaque tranche, ind reçoit le numéro de tranche et tr la tranche en question
    backers = data.loc[indices==ind,"backers"] # sélection des individus de la tranche ind
    if len(backers) > 0:
        g = {
            'valeurs': backers,
            'centre_classe': tr-(bins/2),
            'taille': len(backers),
            'quartiles': [np.percentile(backers,p) for p in [25,50,75]]
        }
        groupes.append(g)

# affichage des boxplots
plt.boxplot([g["valeurs"] for g in groupes],
            positions= [g["centre_classe"] for g in groupes], # abscisses des boxplots
            showfliers= False, # on ne prend pas en compte les outliers
            widths= bins*0.7, # largeur graphique des boxplots
            manage_xticks= False)

# affichage des effectifs de chaque classe
for g in groupes:
    plt.text(g["centre_classe"],0,"(n={})".format(g["taille"]),horizontalalignment='center',verticalalignment='top')     
plt.show()

# affichage des quartiles
for n_quartile in range(3):
    plt.plot([g["centre_classe"] for g in groupes],
             [g["quartiles"][n_quartile] for g in groupes])
plt.show()
```

**ANOVA** : permet de mesurer le **rapport de corrélation** noté $\eta^2$ (eta carré) entre une var quali. et une var quanti. Il est compris entre 0 et 1.

- $\eta^2_{Y/X} = 0$ : pas de corrélation
- $\eta^2_{Y/X} = 1$ : corrélation

```python
X = "main_category" # qualitative
Y = "backers" # quantitative

sous_echantillon = data[data["backers"] >= 0] # On ne garde que les dépenses

def eta_squared(x,y):
    moyenne_y = y.mean()
    classes = []
    for classe in x.unique():
        yi_classe = y[x==classe]
        classes.append({'ni': len(yi_classe),
                        'moyenne_classe': yi_classe.mean()})
    SCT = sum([(yj-moyenne_y)**2 for yj in y])
    SCE = sum([c['ni']*(c['moyenne_classe']-moyenne_y)**2 for c in classes])
    return SCE/SCT
    
eta_squared(sous_echantillon[X],sous_echantillon[Y])
```

Le graphique indiquait des différences en fonction des catégories, mais le rapport de corrélation calculé montre que ces catégories à elles seules ne suffisent absolument pas expliquer le nombre de backers. La corrélation est très faible, d'autres variables entrent en jeu.
